## 1. 

- What is the expected value of a single die roll? 
- What is the expected value of rolling two dice and adding the results together?
- What is the expected winnings of any gamble in European roulette?

- Imagine you roll a die, and you record the value you get. But, if you roll a six, you roll again, and add that value. What is the expected value?
- Imagine that the process described in the last question continues until you fail to roll a six. What is the expected value of the process? (This can be tricky, you can simulate it to get an answer if you prefer. Hint: The answer is 4.2.)

![caption](written_solutions/prob-1.png)

```python
rng = np.random.default_rng(100)

def sim():
    sum = 0
    is_six = 1
    while is_six == 1:
        roll = rng.integers(1,7)
        sum += roll
        if(roll < 6):
            is_six = 0
    return sum

n = 1000000
results = [sim() for i in range(n)]

print(np.mean(results))
```

The simulation reveals that the expected value of the *process* is ~4.2 (compared to less than 4.1 when stopping after two rolls no matter what).

### Bootstrapped density & distribution of $\hat{p}$

```python
y = pd.to_numeric(y.replace({'NO':'0', 'YES':'1'}), errors='coerce')
print(f'MLE: ', y.mean())

S = 5000
sample = [np.mean(y.sample(frac=1,replace=True)) for s in range(S)]

fig, axes = plt.subplots(1,2,figsize=(14,5))
sns.kdeplot(x=sample,ax=axes[0],label='kde')
axes[0].set_title('Bootstrapped KDE of $\\hat{p}$')
sns.ecdfplot(x=sample,ax=axes[1],label='ecdf')
axes[1].set_title('Bootstrapped ECDF of $\\hat{p}$')
plt.show()
```

```python
y = metabric['Mutation Count']
sns.histplot(y, stat='density', bins=45)
```

### MLE derivation
![caption](written_solutions/prob-12.png)

### Model fit

```python
# MLE (sample mean)
lambda_hat = y.mean()
print(lambda_hat)
```

```python
# Integer support
k = np.arange(0, y.max() + 1)

# Plot 1: KDE vs PMF
plt.figure(figsize=(7, 5))

sns.kdeplot(y, bw_adjust=1, label='Sample KDE', linewidth=2)

pmf = sp.stats.poisson.pmf(k, mu=lambda_hat)
plt.stem(k, pmf, basefmt=' ', linefmt='orange', markerfmt='.', label='PMF')

plt.xlabel('Mutation Count')
plt.ylabel('Density')
plt.legend()
plt.show()

# Plot 2: ECDF vs CDF
plt.figure(figsize=(7, 5))

ecdf_x = np.sort(y)
ecdf_y = np.arange(1, len(y)+1) / len(y)
plt.step(ecdf_x, ecdf_y, where='post', label='Sample ECDF')

cdf = sp.stats.poisson.cdf(k, mu=lambda_hat)
plt.step(k, cdf, where='post', color='orange', label='CDF')

plt.xlabel('Mutation Count')
plt.ylabel('Proportion')
plt.legend()
plt.show()
```

Not too shabby... The fitted model predicts slightly lighter tails (i.e., less outliers) than observed in the data, but Poisson seems like a strong fit overall.
